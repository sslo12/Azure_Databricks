```python
# Lista de notebooks 
notebooks_a_ejecutar = [  
"/Workspace/MJCR-MABD-PROD/bronze_ingest",
"/Workspace/MJCR-MABD-PROD/silver_transform", 
"/Workspace/MJCR-MABD-PROD/gold_agg"]

def ejecutar_notebook(notebook, parametros=None):
    try:
        # Agregar el parámetro "environment" a los parámetros del notebook
        parametros = parametros or {}
        parametros["Ambiente"] = environment
        
        print(f"Iniciando ejecución: {notebook} con parámetros: {parametros or '{}'}...")
        resultado = dbutils.notebook.run(notebook, 0, parametros or {})
        print(f"Ejecución completada: {notebook}")
        return resultado  # Puedes capturar el resultado si el notebook devuelve algo
    
    except Exception as e:
        print(f"Error al ejecutar {notebook}: {str(e)}")


# Ejecutar los notebooks con parámetros
for i, nb in enumerate(notebooks_a_ejecutar, start=1):
    print(f"Ejecutando notebook {i}/{len(notebooks_a_ejecutar)}...")
    ejecutar_notebook(nb)

print("Todos los notebooks han sido procesados.")
```
